```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
import datetime
```

```python
x = [datetime.datetime.now(), datetime.datetime.now()+datetime.timedelta(hours=1)]
y = [4, 4]
```

```python
list(x)
```

See: http://nbviewer.jupyter.org/github/WeatherGod/AnatomyOfMatplotlib/blob/master/AnatomyOfMatplotlib-Part1-Figures_Subplots_and_layouts.ipynb#Multiple-Axes

```python
fig, ax = plt.subplots()
ax.plot(x, y);
```

```python
x = [datetime.datetime.now()-datetime.timedelta(hours=1), datetime.datetime.now()+datetime.timedelta(hours=1)]
y = [5, 5]
ax.plot(x, y);
```

```python
xs = zip([datetime.datetime.now()], [datetime.datetime.now()+datetime.timedelta(hours=3)])
ys = zip([7], [7])
for x,y in zip(xs, ys):
    ax.plot(x, y)
```
